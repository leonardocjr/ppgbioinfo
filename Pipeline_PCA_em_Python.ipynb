```python
# import all libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("TCGA-BRCA_MULTI.csv", low_memory=False)
```

```python
df
```

```python
df_pca = df
```

```python
df_pca = df_pca.drop(columns=["Sample"])
```

```python
df_pca = df_pca.drop(columns=["Type"])
```

```python
df_pca
```

```python
# Scale data before applying PCA
scaling=StandardScaler()
 
# Use fit and transform method
scaling.fit(df_pca)
Scaled_data=scaling.transform(df_pca)
 
# Set the n_components=3
principal=PCA(n_components=184)
principal.fit(Scaled_data)
x = principal.transform(Scaled_data)
```

```python
print(x.shape)
```

```python
teste = pd.DataFrame(data=x)
```
