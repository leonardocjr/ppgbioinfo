```python
def find_columns_with_nulls_or_zeros(df):
    """
    Encontra colunas em um DataFrame do Pandas que contêm valores nulos, 0 ou 0.0.
    
    :param df: DataFrame do Pandas
    :return: Lista de nomes de colunas que contêm valores nulos, 0 ou 0.0
    """
    # Criar uma máscara booleana para valores nulos, 0 ou 0.0
    mask = (df.isnull()) | (df == 0) | (df == 0.0)
    
    # Encontrar as colunas que têm pelo menos um valor nulo, 0 ou 0.0
    columns = mask.any().loc[lambda x: x].index.tolist()
    
    return columns
```

```python
columns_to_drop = find_columns_with_nulls_or_zeros(GSE25307)
```

```python
GSE25307_2 = GSE25307.drop(columns_to_drop, axis=1)
```
